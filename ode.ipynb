```python
import numpy as np
from scipy.integrate import odeint
from scipy.optimize import fsolve
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def reaction(X, t):
    Xa = X[0]
    Y = X[1] # Y
    
    dXdt = 1 + 0.02 * np.square(Xa) * Y - 2.04 * Xa
    dYdt = 2 * Xa - 0.02 * np.square(Xa) * Y
    
    return [dXdt, dYdt]

t = np.linspace(0,500,500) # for 500 time steps
X0 = [0,0] # initial concentration = 0
X = odeint(reaction,X0,t)

plt.plot(t,X[:,0],'r--',linewidth=2.0)
plt.plot(t,X[:,1],'b-',linewidth=2.0)
plt.xlabel('Time, t')
plt.ylabel('Concentration')
plt.legend(['X','Y'])
```
